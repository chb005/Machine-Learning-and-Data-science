```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car1=pd.read_csv("ToyotaCorolla.csv",index_col=0,na_values=["??"])
```

```python
car1
```

```python
car1.dropna(axis=0,inplace=True)
```

```python
sns.set(style="darkgrid")
sns.regplot(x=car1["Age_08_04"],y=car1["Price"],fit_reg=False,marker="*")
```

```python
sns.lmplot(x='Age_08_04',y='Price',data=car1,fit_reg=False,hue='Fuel_Type',legend=True,palette='Set1')
```

```python
sns.distplot(car1["Age_08_04"],kde=False,bins=5)
```

```python
sns.countplot(x='Fuel_Type',data=car1,hue="Automatic")
```

```python
sns.boxplot(y=car1['Price'])
```

```python
sns.boxplot(x=car1['Fuel_Type'],y=car1['Price'],hue='Automatic',data=car1)
```

```python
sns.pairplot(car1,kind="scatter",hue='Fuel_Type')
plt.show()
```
