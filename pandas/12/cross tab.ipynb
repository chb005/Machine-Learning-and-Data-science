```python
import pandas as pd
```

```python
df=pd.read_excel("survey.xls")
```

```python
pd.crosstab(df.Nationality,df.Handedness)
```

```python
pd.crosstab(df.Sex,df.Handedness)
```

```python
pd.crosstab([df.Sex,df.Nationality],df.Handedness)
```

```python
pd.crosstab(df.Sex,[df.Handedness,df.Nationality],margins=False,normalize=False)
```

```python
pd.crosstab(df.Sex,[df.Handedness,df.Nationality],normalize=True)
```

```python
import numpy as np
pd.crosstab(df.Sex,[df.Handedness,df.Nationality],values=df.Age,aggfunc=np.average,normalize=True)
```
