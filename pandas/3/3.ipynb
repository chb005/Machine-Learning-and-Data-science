```python
df=pd.read_excel("stock_data.xlsx",sheet_name="Sheet2")
df
```

```python
def convert_na(cell):
    if cell=="n.a.":
        return "chirag"
    return cell
```

```python
df=pd.read_excel("stock_data.xlsx","Sheet1",converters={
    'people':convert_na
})
df
```

```python
df.to_excel("chb2.xlsx",sheet_name="chirag1",index=False,startcol=1,startrow=2)
```
